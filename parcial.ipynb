```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos originales
servicios = ['Consulta general', 'Vacunación', 'Desparasitación',
             'Cirugía menor', 'Radiografía', 'Hospitalización']
precios = [75325, 58503, 101843, 94004, 98456, 115930]

# Crear dataframe
df = pd.DataFrame({
    'ID_Servicio': [1, 2, 3, 4, 5, 6],
    'Nombre': servicios,
    'Precio': precios
})

# Datos ficticios de frecuencia por tipo de mascota (simulados)
perros = [120, 100, 80, 50, 40, 30]
gatos = [90, 85, 60, 40, 45, 25]
aves = [30, 25, 10, 15, 20, 5]
reptiles = [15, 10, 20, 5, 25, 8]

# 1. Gráfico de barras - el más simple
plt.figure(figsize=(10, 6))
plt.bar(servicios, perros, label='Perros')
plt.bar(servicios, gatos, bottom=perros, label='Gatos')
plt.bar(servicios, aves, bottom=[p+g for p,g in zip(perros,gatos)], label='Aves')
plt.bar(servicios, reptiles, bottom=[p+g+a for p,g,a in zip(perros,gatos,aves)], label='Reptiles')

plt.xlabel('Servicios')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Servicios por Tipo de Mascota')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig('barras_simple.png')
plt.show()  # Mostrar gráfico de barras

# 2. Gráfico de dispersión simple
plt.figure(figsize=(8, 5))
plt.scatter(precios, perros, label='Perros', s=100)
plt.scatter(precios, gatos, label='Gatos', s=100)
plt.scatter(precios, aves, label='Aves', s=100)
plt.scatter(precios, reptiles, label='Reptiles', s=100)

plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Precio vs Frecuencia')
plt.legend()
plt.grid(True)
plt.savefig('dispersion_simple.png')
plt.show()  # Mostrar gráfico de dispersión

# 3. Histograma muy básico
plt.figure(figsize=(8, 5))
plt.hist(perros, alpha=0.5, label='Perros')
plt.hist(gatos, alpha=0.5, label='Gatos')
plt.hist(aves, alpha=0.5, label='Aves')
plt.hist(reptiles, alpha=0.5, label='Reptiles')

plt.xlabel('Frecuencia')
plt.ylabel('Conteo')
plt.title('Distribución de Frecuencias')
plt.legend()
plt.savefig('histograma_simple.png')
plt.show()  # Mostrar histograma

# Mostrar información del dataframe
df.info()
```
