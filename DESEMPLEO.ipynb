# Tasa de desempleo por sexo a lo largo de los años

# Hipotesis

 * Existe una persistente brecha de género en el desempleo durante todo el período analizado (2015-2024), siendo sistemáticamente mayor la tasa de desempleo de las mujeres.


```python
# Asegurarse de que la fecha sea tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Año'] = df['Fecha'].dt.year

# Filtrar solo personas desempleadas
desempleados = df[df['Estado_Laboral'] == 'Desempleado']

# Total por año y sexo
totales = df.groupby(['Año', 'Sexo']).size().rename('Total')
desemp = desempleados.groupby(['Año', 'Sexo']).size().rename('Desempleados')

# Unir y calcular tasa de desempleo
tasa = pd.concat([totales, desemp], axis=1)
tasa['Tasa_Desempleo'] = (tasa['Desempleados'] / tasa['Total']) * 100
tasa = tasa.reset_index()

# Gráfico
plt.figure(figsize=(10, 6))
sns.lineplot(data=tasa, x='Año', y='Tasa_Desempleo', hue='Sexo', marker='o')
plt.title('Tasa de Desempleo por Sexo a lo Largo de los Años')
plt.ylabel('Tasa de Desempleo (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig('Tasa de Desempleo por Sexo a lo Largo de los Años.png')
plt.show()
```



1.   Gráfico:
Líneas de tendencia por año, una para hombres y otra para mujeres, mostrando la evolución de la tasa de desempleo anual.

2.   Análisis:


*   Se observa un aumento brusco en 2020 (año de la pandemia) para ambos sexos, más pronunciado en mujeres.
*   Aunque las tasas bajan en 2022 y 2023, las mujeres siguen teniendo mayor desempleo en todos los años.

* La brecha de género se acentúa durante los años más críticos (2020–2021).


3. Conclusión: El impacto de la pandemia fue más severo para las mujeres. Aunque hay una recuperación posterior, la desigualdad persiste. Esto refleja que las crisis económicas agravan las brechas existentes si no hay medidas correctivas específicas.

# Dispersión de la Tasa de Desempleo Mensual en Colombia (2015–2024)


image.png

image.png


image.png
Hipotesis


*  Se observa un punto extremadamente alto (outlier) en 2020, con una tasa de desempleo cercana al 38%, que probablemente coincide con el impacto inicial de la pandemia de COVID-19.




```python
# Filtrar solo personas en la fuerza laboral (empleados y desempleados)
fuerza_laboral = df[df['Estado_Laboral'].isin(['Empleado', 'Desempleado'])].copy()

# Crear columna Año-Mes para agrupar
fuerza_laboral['AñoMes'] = fuerza_laboral['Fecha'].dt.to_period('M')

# Calcular conteos por mes y estado laboral
desempleo_mensual = (
    fuerza_laboral.groupby('AñoMes')['Estado_Laboral']
    .value_counts()
    .unstack()
    .fillna(0)
)

# Calcular la tasa de desempleo
desempleo_mensual['Tasa_Desempleo'] = (
    desempleo_mensual['Desempleado'] /
    (desempleo_mensual['Desempleado'] + desempleo_mensual['Empleado']) * 100
)

# Convertir 'AñoMes' a una fecha (primer día del mes) para graficar
desempleo_mensual['Fecha'] = desempleo_mensual.index.to_timestamp()

# Gráfica de dispersión
plt.figure(figsize=(16, 8))
sns.scatterplot(data=desempleo_mensual, x='Fecha', y='Tasa_Desempleo', color='darkred')

plt.title('Dispersión de la Tasa de Desempleo Mensual en Colombia (2015–2024)', fontsize=14)
plt.xlabel('Fecha')
plt.ylabel('Tasa de Desempleo (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig('dispersion_tasa_desempleo_mensual.png')
plt.show()
```
